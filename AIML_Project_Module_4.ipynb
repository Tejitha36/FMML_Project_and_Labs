## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # Calculate the gradient (derivative of f(x) at new_x)
    new_x = previous_x - step_multiplier * gradient
    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # Append the new gradient value to slope_list

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
## TASK-7 : Plot reduction in cost for each iteration
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    """
    Gradient Descent function to find local minimum.

    Args:
        derivative_func: Derivative of the function.
        initial_guess: Initial value.
        multiplier: Learning rate.
        precision: Stopping criterion.
        max_iter: Maximum number of iterations.

    Returns:
        new_x, x_list, slope_list: Local minima, x values and derivative at x values
    """
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
  # Define the cost function
def cost_function(x):
    return x**2

# Now you can use it to compute costs for different learning rates
n = 100  # Maximum iterations
low_gamma_result, low_gamma_x_list, _ = gradient_descent(derivative_func=dg, initial_guess=3,
                                                         multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma_result, mid_gamma_x_list, _ = gradient_descent(derivative_func=dg, initial_guess=3,
                                                         multiplier=0.001, precision=0.0001, max_iter=n)

high_gamma_result, high_gamma_x_list, _ = gradient_descent(derivative_func=dg, initial_guess=3,
                                                           multiplier=0.002, precision=0.0001, max_iter=n)

insane_gamma_result, insane_gamma_x_list, _ = gradient_descent(derivative_func=dg, initial_guess=1.9,
                                                               multiplier=0.25, precision=0.0001, max_iter=n)

# Compute costs for each iteration
low_gamma_costs = [cost_function(x) for x in low_gamma_x_list]
mid_gamma_costs = [cost_function(x) for x in mid_gamma_x_list]
high_gamma_costs = [cost_function(x) for x in high_gamma_x_list]
insane_gamma_costs = [cost_function(x) for x in insane_gamma_x_list]

# Plotting the reduction in cost for each learning rate
plt.figure(figsize=(12, 8))
plt.plot(low_gamma_costs, label='Low Gamma (0.0005)', color='blue')
plt.plot(mid_gamma_costs, label='Mid Gamma (0.001)', color='orange')
plt.plot(high_gamma_costs, label='High Gamma (0.002)', color='green')
plt.plot(insane_gamma_costs, label='Insane Gamma (0.25)', color='red')

# Adding labels and title
plt.title('Reduction in Cost per Iteration for Different Learning Rates')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()
plt.grid()
plt.show()
```

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples

# Define the function f(x, y)
def f(x, y):
    return 1 / (3**(-x**2 - y**2) + 1)
# Define the partial derivative with respect to x
def df_dx(x, y):
    r = 3**(-x**2 - y**2) + 1
    return -(1 / r**2) * (2 * x * np.log(3) * 3**(-x**2 - y**2))

# Define the partial derivative with respect to y
def df_dy(x, y):
    r = 3**(-x**2 - y**2) + 1
    return -(1 / r**2) * (2 * y * np.log(3) * 3**(-x**2 - y**2))

# Gradient Descent Parameters
initial_x = 1.0  # Starting point for x
initial_y = 1.0  # Starting point for y
step_multiplier = 0.01  # Learning rate
precision = 0.00001

# Lists to store values for plotting and analysis
x_list = [initial_x]
y_list = [initial_y]

# Gradient Descent Loop
for n in range(500):
    gradient_x = df_dx(initial_x, initial_y)  # Get the gradient with respect to x
    gradient_y = df_dy(initial_x, initial_y)  # Get the gradient with respect to y

    initial_x -= step_multiplier * gradient_x  # Update x using gradient descent
    initial_y -= step_multiplier * gradient_y  # Update y using gradient descent

    x_list.append(initial_x)
    y_list.append(initial_y)

    if np.sqrt(gradient_x**2 + gradient_y**2) < precision:
        print('Loop ran this many times:', n)
        break

# Output results
print('Local minimum occurs at (x, y):', (initial_x, initial_y))
print('Function value at this point is:', f(initial_x, initial_y))
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
